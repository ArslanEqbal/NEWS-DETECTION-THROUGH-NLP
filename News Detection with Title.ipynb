```python
import nltk
import re
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
```

```python
ps = PorterStemmer()
corpus = []


for i in range(0,len(df)):
    review = re.sub('[^a-zA-Z]',' ',df['title'][i])
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word)for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
import tensorflow as tf

from tensorflow.keras.layers import Embedding
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dropout
```

```python
vocabulary_size = 5000
```

```python
one_hot_representation = [one_hot(words,vocabulary_size) for words in corpus]
```

```python
one_hot_representation
```

## MODELLING

##### Starting with embedding with over 50 dimensions and a fixed sentence length of 20 words

```python
sentence_length = 20
embedded_values = pad_sequences(one_hot_representation , padding = 'pre' , maxlen = sentence_length)
```

```python
embedded_values
```

```python
len(embedded_values)
```

BUILDING MODEL

```python
embedded_vector_features = 40
model = Sequential()
model.add(Embedding(vocabulary_size , embedded_vector_features , input_length = sentence_length))
model.add(Dropout(0.3))
model.add(LSTM(100))
model.add(Dropout(0.3))
model.add(Dense(1 , activation = 'sigmoid'))
model.compile(loss = 'binary_crossentropy' , optimizer = 'adam' , metrics  =['accuracy'])
model.summary()
```
